```python
texts=[]
with open('D:/mycode/NLP/keyword/kr-wordrank/data/134963.txt', "r", encoding='utf-8') as infile:
    for line in infile:
        a,b=line.split('\t')
        texts.append(a)
texts[:10]     
```

```python
from textrank import KeywordSummarizer
from konlpy.tag import Komoran
from konlpy.tag import Okt
import re
def regex(x):
    
    x=re.sub("\[.*\]|\s-\s.*",'',x)
    x=re.sub("\(.*\)|\s-\s.*",'',x)
    x=re.sub("\<.*\>|\s-\s.*",'',x)
    x=re.sub("\【.*\】|\s-\s.*",'',x)
    x= re.sub('[-=+#/\?:^$@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》,▷▶ⓒ]', '', x)
    
    return x

texts_re=[]
for i in texts:
    a=regex(str(i))
    texts_re.append(a)

komoran = Komoran()
okt = Okt()

def remove(x): 
    x= re.sub('[/NNG,/NNP,/Noun]', '', x)
    return x
    

def komoran_tokenizer(sent):
    words = komoran.pos(sent, join=True)
    words = [remove(w) for w in words if ('/NNG' in w or '/NNP' in w)]
    return words

def okt_tokenizer(sent):
    words = okt.pos(sent, join=True)
    words = [remove(w) for w in words if ('/Noun' in w) ]
    return words

keyword_extractor_komoran = KeywordSummarizer(
    tokenize = komoran_tokenizer,min_count=2, min_cooccurrence=1,
    window=-1,                
    verbose = False)

keyword_extractor_okt = KeywordSummarizer(
    tokenize = okt_tokenizer,min_count=2, min_cooccurrence=1,
    window=-1,                
    verbose = False)
```

```python
# komoran 형태소 명사키워드
keywords = keyword_extractor_komoran.summarize(texts_re, topk=100)
keywords[:20]
```

```python
# okt 형태소 키워드
keywords_okt = keyword_extractor_okt.summarize(texts, topk=100)
keywords_okt[:20]
```

```python
words=[]
rank=[]
for word,_ in keywords:
    words.append(word)
    rank.append(_)
    
import pandas as pd
df_komoran=pd.DataFrame({'keys':words,'rank':rank})
df_komoran[:10]
```

```python
words1=[]
rank1=[]
for word,_ in keywords_okt:
    words1.append(word)
    rank1.append(_)
    
import pandas as pd
df_okt=pd.DataFrame({'keys':words1,'rank':rank1})
df_okt[:10]
```
