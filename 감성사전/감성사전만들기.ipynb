```python
import pandas as pd
df=pd.read_csv('D:/크롤링/anbotrain6_8eda.csv')
df
```

```python
from konlpy.tag import Okt
import numpy as np
import re
okt = Okt()
n_ = []
title_rename = []
for i in range(len(df)):
    if(i % 10000 ==0):
        print(i,"단계 완료")
    title_rename.append(re.sub("[\(\[].*?[\)\]]", "", df.iloc[i]['texts']))
    n_.append(' '.join(okt.nouns(df.iloc[i]['texts'])))
df['nouns'] = n_
df['texts']=title_rename
df = df[df['nouns']!='']

#2차 불용어 제거
df['texts'] = df['texts'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","")
df['texts'].replace('', np.nan, inplace=True)
df = df.dropna(how='any')
```

```python
vocab={}
cnt=0
for i in df['nouns']:
    i=i.split(' ')
    for j in range(len(i)):
        if i[j] in vocab or len(i[j])<=1:
            cnt=cnt+1
            pass
        else:
            vocab[i[j]] =0
vocab
print(len(vocab),vocab)
```

```python
up = 17076+18286
down = 35078
up_ratio = up/(len(df))
down_ratio = down/(len(df))

import collections
for i,w in enumerate(df['nouns']):
    w = w.split(' ')
    if (df.iloc[i]['labels']==1):
        for j in range(len(w)):
            noun = w[j]
            if len(noun)<=1:
              continue
            vocab[noun] = vocab[noun] + up_ratio
    else:
        for j in range(len(w)):
            noun = w[j]
            if len(noun)<=1:
              continue
            vocab[noun] = vocab[noun] - down_ratio
```

```python
total = []
sent_dictionary=vocab
for i,w  in enumerate(df['nouns']):
    sent_score = 0
    w= w.split(' ')
    for j in w:
        if(len(j)<=1):
          continue
        elif(j not in sent_dictionary):
          continue
        else:
          sent_score = sent_score + sent_dictionary[j]
    total.append(sent_score/len(w))
df['sent_score'] = total
```
